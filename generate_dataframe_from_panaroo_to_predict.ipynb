```python
import pandas as pd
```

```python
df = pd.read_csv("gene_presence_absence.Rtab", sep = "\t", header=0, usecols=['Gene', "x"])
```

```python
df = df.set_index("Gene").transpose()
df
```

```python
df = df.reset_index().rename(columns={'index': 'Gene'})
df
```

```python
col_keep = ['Gene', 'Annotation']
df2 = pd.read_csv("gene_presence_absence.csv", header=0, low_memory=False, usecols=col_keep)
df2.shape
```

```python
locus_tag = df2["Annotation"]
locus_tag = locus_tag.str.split('locus_tag=', 1 , expand=True)
locus_tag = locus_tag[1]
locus_tag = locus_tag.str.split(']', 1 , expand=True).fillna('Empty')
locus_tag = locus_tag[0].tolist()
```

```python
key_list = list(df2['Gene'])
dict_lookup = dict(zip(df2['Gene'], locus_tag))
```

```python
red = df.columns.values
```

```python
red[1:] = [dict_lookup[item] for item in key_list]
print(red)
df.columns = red
df.head(5)
```

```python
df = df.groupby(by=df.columns, axis=1).sum()
df = df.drop(["Empty"], axis = 1)
df.shape
df
```

```python
list_features = pd.read_csv("list_of_features_model_A.txt", header=None).T
list_features.columns = list_features.iloc[0]

list_features
```

```python
list_features2 = pd.read_csv("list_of_features_model_B.txt", header=None).T
list_features2.columns = list_features2.iloc[0]

list_features2
```

```python
list_features3 = pd.read_csv("list_of_features_model_C.txt", header=None).T
list_features3.columns = list_features3.iloc[0]

list_features3
```

```python
df2 = pd.concat([list_features, df], join = "inner")
df2 = df2.reset_index()
df2
```

```python
df3 = pd.concat([list_features2, df], join = "inner")
df3 = df3.reset_index()
df3
```

```python
df4 = pd.concat([list_features3, df], join = "inner")
df4 = df4.reset_index()
df4
```

```python
df2 = df2.drop([0])
df2
```

```python
df3 = df3.drop([0])
df3
```

```python
df4 = df4.drop([0])
df4
```

```python
df2 = pd.concat([df2, list_features], join = "outer").fillna(0)
df2 = df2.drop([0])
df2
```

```python
df3 = pd.concat([df3, list_features2], join = "outer").fillna(0)
df3 = df3.drop([0])
df3
```

```python
df4 = pd.concat([df4, list_features3], join = "outer").fillna(0)
df4 = df4.drop([0])
df4
```

```python
df2 = df2.rename(columns={"index" : "Gene"})
```

```python
df3 = df3.rename(columns={"index" : "Gene"})
```

```python
df4 = df4.rename(columns={"index" : "Gene"})
```
